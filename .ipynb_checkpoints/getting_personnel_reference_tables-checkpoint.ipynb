```python
def make_manual_date_ref():
    dates_ref_list = [
        ['2022-08-01','2022-08-31','2022-august'],
        ['2022-09-01','2022-09-30','2022-september'],
        ['2022-10-01','2022-10-31','2022-october'],
        ['2022-11-01','2022-11-30','2022-november'],
        ['2022-12-01','2022-12-31','2023-december'],
        ['2023-01-01','2023-01-31','2023-january'],
        ['2023-02-01','2023-02-28','2023-february'],
        ['2023-03-01','2023-03-31','2023-march'],
        ['2023-04-01','2023-04-30','2023-april'],
        ['2023-05-01','2023-05-31','2023-may'],
        ['2023-06-01','2023-06-30','2023-june'],
    ]

    df_dates_ref = pd.DataFrame(dates_ref_list, columns=['first_of_month','end_of_month','year_month'])

    df_dates_ref['first_of_month'] = pd.to_datetime(df_dates_ref['first_of_month'])
    df_dates_ref['end_of_month'] = pd.to_datetime(df_dates_ref['end_of_month'])
    
    return df_dates_ref
```

```python
def get_daily_date_range(date1, date2):
    date_range = pd.date_range(date1, date2)
    return
```

```python
def get_unassigned_tasks_by_role(df, role):
    df = df[df['role'] == role]
    
    if role == 'Commissioning Manager':
        df = df[df['assignee_name'] == 'Antti Pitkänen']
    else:
        df = df[df['assignee_name']=='unassigned'] 
    return df
```

```python
def remove_unassigned_tasks_by_role(df, role):
    df = df[df['role'] == role]
    
    if role == 'Commissioning Manager':
        df = df[df['assignee_name'] != 'Antti Pitkänen']
        df = df[df['assignee_name'] !='unassigned']
    else:
        df = df[df['assignee_name'] !='unassigned']
    return df
```

```python
def get_num_active_pjs(df, start_date, end_date):
    df_out = wa.filter_tasks_by_time_window(df, start_date, end_date)['eso_project_name']
    pj_count = df_out.nunique()
    return pj_count
```

```python
test_list = [{
    'month' : r['year_month'],
    'num_projects_needing_cx_mgr': get_num_active_pjs(df_cx_upcoming, r['first_of_month'], r['end_of_month']),
    'num_active_cx_mgrs': get_num_active_cxs(df_cx_assigned, r['first_of_month'], r['end_of_month']), 
    } for _, r in df_dater.iterrows()]
```

```python
def make_supply_status_cx_mgr(row):
    supply = row['num_cx_managers_available']
    need = row['num_projects_needing_cx_mgr']
    if need > supply: 
        status = 'red'
    elif need == supply: 
        status = 'yellow'
    else: 
        status = 'green'
    return status
    
```

```python
df_cx_analysis = pd.DataFrame(test_list)
total_employed_cx_amer = 3
df_cx_analysis['num_cx_managers_available'] = total_employed_cx_amer - df_cx_analysis['num_active_cx_mgrs']
df_cx_analysis['cx_supply_status'] = df_cx_analysis.apply(make_supply_status_cx_mgr, axis=1)

df_cx_analysis
```

```python
df_cx_analysis.plot(x="month", y=['num_projects_needing_cx_mgr', 'num_cx_managers_available'], kind="bar", title='Cx Manager Need vs Supply')
```
