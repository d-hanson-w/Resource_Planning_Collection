```python
def get_unassigned_tasks_by_role(df, role):
    df = df[df['role'] == role]
    
    if role == 'Commissioning Manager':
        df = df[df['assignee_name'] == 'Antti Pitkänen']
    else:
        df = df[df['assignee_name']=='unassigned']
        
    return df
```

```python
def remove_unassigned_tasks_by_role(df, role):
    df = df[df['role'] == role]
    
    if role == 'Commissioning Manager':
        df = df[df['assignee_name'] != 'Antti Pitkänen']
        df = df[df['assignee_name'] !='unassigned']
    else:
        df = df[df['assignee_name'] !='unassigned']
    return df
```

```python
def check_dates_overlap(dt_1_start, dt_1_end, dt_2_start, dt_2_end): 
    if (dt_1_start <= dt_2_end) & (dt_1_end >= dt_2_end): 
        return True
    else: 
        return False
```
