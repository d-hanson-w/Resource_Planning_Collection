## Local Utility Functions

```python
def get_num_unique_pjs_by_window(df, start_date, end_date):
    df_out = wa.filter_tasks_by_time_window(df, start_date, end_date)['eso_project_name']
    pj_count = df_out.nunique()
    return pj_count
```

```python
def get_list_unique_pjs_by_window(df, start_date, end_date):
    df_out = wa.filter_tasks_by_time_window(df, start_date, end_date)['eso_project_name']
    pj_list = df_out.unique().tolist()
    return pj_list
```

# Project Analysis

```python
def get_projects_count(df_in):
    start = pd.Timestamp('1/1/2022')
    end = pd.Timestamp('1/1/2024')
    tindex = pd.date_range(start, end)
    df_daily = pd.DataFrame(tindex, columns=['date'])

    data_list = [{
       'date':r['date'],
        'num_active_projects' : get_num_unique_pjs_by_window(df_in, r['date'], r['date'])
    } for _, r in df_daily.iterrows()]

    return pd.DataFrame(data_list)
```

```python
def get_projects_count_by_region(df_in):
    start = pd.Timestamp('1/1/2022')
    end = pd.Timestamp('1/1/2024')
    tindex = pd.date_range(start, end)
    df_daily = pd.DataFrame(tindex, columns=['date'])

    region_list = ['AMER', 'EUAF', 'MEA']
    
    data_list = list()
    
    for _, r in df_daily.iterrows():
        win_pjs = wa.filter_tasks_by_time_window(df_in, r['date'], r['date'])
        win_dict = {'date' : r['date']}
        for reg in region_list:
            win_dict[reg] = win_pjs[win_pjs['region'] == reg]['eso_project_name'].nunique()
        
        win_dict['ALL'] = win_dict['AMER'] + win_dict['EUAF'] + win_dict['MEA']
        data_list.append(win_dict)
        
    return pd.DataFrame(data_list)
```

```python
df_tasks = df_all.copy()
px.line(get_projects_count(df_tasks), x='date', y='num_active_projects')
```

```python
df_tasks = df_all.copy()
selected_regions = ['AMER', 'EUAF', 'MEA', 'no_region']
df_tasks = df_tasks[df_tasks['region'].isin(selected_regions)]
px.line(get_projects_count(df_tasks), x='date', y='num_active_projects')
```

```python
def make_project_count_graph_by_region(df_tasks):
    df_pj_reg_count = get_projects_count_by_region(df_tasks)

    df_melt_count = pd.melt(df_pj_reg_count, id_vars='date', value_vars=['AMER', 'EUAF', 'MEA','ALL'])
    df_melt_count = df_melt_count.rename({'date':'date', 'variable':'region','value':'project_count'}, axis=1)
    fig = px.line(df_melt_count, x='date', y='project_count', color='region')
    return fig
```

```python
df_tasks = df_all.copy()
make_project_count_graph_by_region(df_tasks).show()
```

# Masking Data for Analysis

```python
def get_role_reference_df(roles=['all']): 
    """
    Parameters
    ----------------
    df : pandas dataframe
        containing all active employees in Wartsila ESO Delivery and Commissioning Teams
    Returns
    ----------------
    out_dict : dict
        count :  number of people in each role
        list :   each person in each role
    """
    # get data from asana
    pdb_list = wa.get_persondb_api_data()
    df = wa.make_persondb_df(pdb_list)
    # filter by selected roles
    if roles == ['all']:
        return df
    else: 
        df = df[df['role'].isin(roles)].sort_values('role')
        out_dict = {r : {'count':df[df['role']==r]['person_name'].nunique(),
                      'list': df[df['role']==r]['person_name'].tolist() 
                      } for r in roles}
        
    return out_dict
```
